# Tampilan Data Capture

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy as scipy
from scipy import optimize
from matplotlib.ticker import AutoMinorLocator
from matplotlib import gridspec
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
# numerical data file
# sampling rate: 10 ms
# capture 1 sampai 3: gerakan cepat dan step-step
filename1="capture1.txt" #tanpa kapasitor
filename2="capture2.txt" #kapasitor 100 nF
filename3="capture3.txt" #kapasitor 10uF

# capture 4 sampai 6: gerakan lambat dan cepat
filename4="capture4.txt" #tanpa kapasitor
filename5="capture5.txt" #tanpa kapasitor
filename6="capture6.txt" #kapasitor 10uF
```

```python
data1 = np.loadtxt(filename1, delimiter=",")
data2 = np.loadtxt(filename2, delimiter=",")
data3 = np.loadtxt(filename3, delimiter=",")
data4 = np.loadtxt(filename4, delimiter=",")
data5 = np.loadtxt(filename5, delimiter=",")
data6 = np.loadtxt(filename6, delimiter=",")
```

```python
fig = plt.figure(figsize=(16,4))
gs = gridspec.GridSpec(1,1)
ax1 = fig.add_subplot(gs[0])
ax1.plot(data1, "r-")
ax1.set_xlim(0,2500)
fig.savefig("data1.png", format="png",dpi=60)
```
