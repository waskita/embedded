# Tampilan Data Capture

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy as scipy
from scipy import optimize
from matplotlib.ticker import AutoMinorLocator
from matplotlib import gridspec
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
# numerical data file
# sampling rate: 10 ms
# capture 1 sampai 3: gerakan cepat dan step-step
filename0="capture.txt" #tanpa kapasitor
filename1="capture1.txt" #tanpa kapasitor
filename2="capture2.txt" #kapasitor 100 nF
filename3="capture3.txt" #kapasitor 10uF
filename4="capture4.txt" # tegangan input 12 volt, ada masalah dengan serial output
filename5="capture5.txt" 
filename6="capture6.txt" 
filename7="c:/temp/capture.txt"  # 14 volt, dengan software baru
# capture 4 sampai 6: gerakan lambat dan cepat


```

```python
#data0 = np.loadtxt(filename0, delimiter=",")
data1 = np.loadtxt(filename1, delimiter=",")
data2 = np.loadtxt(filename2, delimiter=",")
data3 = np.loadtxt(filename3, delimiter=",")
data4 = np.loadtxt(filename4, delimiter=",")
data5 = np.loadtxt(filename5, delimiter=",")
data6 = np.loadtxt(filename6, delimiter=",")
data7 = np.loadtxt(filename7, delimiter=",")
```

```python
fig = plt.figure(figsize=(16,4))
gs = gridspec.GridSpec(1,1)
ax1 = fig.add_subplot(gs[0])
#ax1.plot(data1, "r-")
#ax1.plot(data3, "b-")
ax1.plot(data5, "m-", label="input 4 volt")
ax1.plot(data2, "g-", label="input 8 volt")
ax1.plot(data4, "c-", label="input 12 volt")
ax1.plot(data6, "r-", label="input 14 volt")
ax1.plot(data7, "b-", label="input 13 volt, new")
#ax1.set_xlim([0,500])
ax1.legend()
fig.savefig("gabungan-beban1.png", format="png",dpi=60)
```

```python
fig = plt.figure(figsize=(16,4))
gs = gridspec.GridSpec(1,1)
ax1 = fig.add_subplot(gs[0])
ax1.plot(data7, "b-", label="input 13 volt, new")
#ax1.set_xlim([0,500])
ax1.legend()
fig.savefig("kurva-13v.png", format="png",dpi=60)
```
