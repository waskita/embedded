```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator
from matplotlib import gridspec
import matplotlib.ticker as ticker
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('win10-filter.csv')
print(df.to_string()) 
```

```python
figure, axis = plt.subplots(2, 1) 
## gs = gridspec.GridSpec(1,1)
ax1 = axis[0]
ax1.plot(df['t'],df['xn'],"r-",label="error")
ax1.plot(df['t'],df['yn'],"g-",label="velocity")
# ax1.plot(df['t'],df['posisi'],"g-",label="yn")

## ax1.plot(1/10,.632,"bo",label="output y")
ax1.set_xlabel("waktu",family="serif",  fontsize=12)
ax1.set_ylabel("value",family="serif",  fontsize=12)
ax1.legend(loc='best')

ax2 = axis[1]
ax2.plot(df['t'],df['setpoint'],"r-",label="setpoint")
ax2.plot(df['t'],df['posisi'],"b-",label="posisi")
ax2.legend(loc='best')

figure.tight_layout()
plt.grid(True)
```
